```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

#### ***Generate sample data***

```python
np.random.seed(0) #Used for same random numbers every time
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
```

## ***Apply Linear Regression***

#### ***Fit Linear Regression model***

```python
reg = LinearRegression()
reg.fit(x, y)
y_pred = reg.predict(x)
```

#### ***Plot the data and best fit line***

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_pred, color='red', linewidth=2, label='Best Fit Line')
plt.title("Linear Regression Best Fit Line")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.grid(True)
plt.show()
```

```python
print("m: ",reg.coef_)
print("c: ",reg.intercept_)
```

## ***Apply Gradient Descent***

```python
#Step-1:
m = 0
c = 0
Learning_rate = 0.01

n = int(len(x))
iterations = 1000

#Step-2:
for i in range(iterations):
    y_prediction = m*x + c
    derivatives_m = (-2/n) * sum(x*(y-y_prediction))
    derivatives_c = (-2/n) * sum(y-y_prediction)

    #Step-3:
    m =  m - Learning_rate * derivatives_m
    c =  c - Learning_rate * derivatives_c

print("m: ",m)
print("c: ",c)
```

##### ***Predicted y using gradient descent***

```python
y_pred_gd = m*x + c
```

#### ***Plot the data and best fit line***

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', alpha=0.6, label='Data Points')
plt.plot(x, y_pred_gd, color='orange', linewidth=2, label='Gradient Descent Line')
plt.title("Best Fit Line using Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")

plt.legend()
plt.grid(True)
plt.show()
```

## ***Gradient Descent Progression***   
## ***Gradually reach to the Minimum point with 1000 iteration***

```python
#Step-1:
m = 0
c = 0
Learning_rate = 0.01

n = int(len(x))
iterations = 1000

# Store all (m, c) after each iteration
history = []

#Step-2:
for i in range(iterations):
    y_prediction = m*x + c
    derivatives_m = (-2/n) * sum(x*(y-y_prediction))
    derivatives_c = (-2/n) * sum(y-y_prediction)

    #Step-3:
    m =  m - Learning_rate * derivatives_m
    c =  c - Learning_rate * derivatives_c
    history.append((m, c))  # Save the values for plotting

#Plot
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', alpha=0.5, label='Data Points')
plt.plot(x, y_pred, color='green', linewidth=2, label='Best Fit Line') #Linear Regression Line
# Plot line at each iteration
for i, (m, c) in enumerate(history):
    y_pred = m * x + c
    if i == len(history) - 1:
        plt.plot(x, y_pred, color='red', linewidth=2.5, label=f'Final Line (Iter {i+1})')
    else:
        plt.plot(x, y_pred, color='orange', alpha=0.3)

plt.title("Gradient Descent Progression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

## ***Gradually reach to the Minimum point with 200 iteration***

```python
def compute_cost(X, y, m, b):
    y_pred = m * X + b
    return np.mean((y - y_pred) ** 2)

# Step-1:
m = 0
c = 0
Learning_rate = 0.01

n = int(len(x))
iterations = 200

# Store all (m, c) after each iteration
history = []

# Step-2:
for i in range(iterations):
    y_prediction = m*x + c
    derivatives_m = (-2/n) * sum(x*(y-y_prediction))
    derivatives_c = (-2/n) * sum(y-y_prediction)

    # Step-3:
    m =  m - Learning_rate * derivatives_m
    c =  c - Learning_rate * derivatives_c
    cost = compute_cost(x, y, m, c)
    history.append((m, c, cost))  # Save the values for plotting

# Extract values for plotting
m_vals = [h[0] for h in history]
c_vals = [h[1] for h in history]
cost_vals = [h[2] for h in history]

# Plot
plt.figure(figsize=(10, 5))

# Plot 1: Gradient Descent Line Progression
plt.subplot(1, 3, 1)
plt.scatter(x, y, color='blue', alpha=0.5, label='Data Points')
plt.plot(x, y_pred, color='green', linewidth=2, label='Best Fit Line')
for i in range(iterations):
    y_pred = m_vals[i] * x + c_vals[i]
    if i == iterations - 1:
        plt.plot(x, y_pred, color='red', linewidth=2, label='Final Fit')
    else:
        plt.plot(x, y_pred, color='orange', alpha=0.3)
plt.title("Gradient Descent Line Progression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Plot 2: Cost vs m
plt.subplot(1, 3, 2)
plt.plot(m_vals, cost_vals, marker='o', color='green')
plt.title("Cost vs m")
plt.xlabel("m")
plt.ylabel("Cost")
plt.grid(True)

# Plot 3: Cost vs b
plt.subplot(1, 3, 3)
plt.plot(c_vals, cost_vals, marker='o', color='purple')
plt.title("Cost vs b")
plt.xlabel("b")
plt.ylabel("Cost")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## ***Increased Learning Rate (L=1)***
##### ***`Small learning rate` → moves down slowly, takes longer to reach the correct solution (minima).***               
##### ***`Large learning rate` → takes bigger jumps, can reach faster, but if too large it may overshoot the solution and keep bouncing without reaching it.***

```python
def compute_cost(X, y, m, b):
    y_pred = m * X + b
    return np.mean((y - y_pred) ** 2)

# Step-1:
m = 0
c = 0
Learning_rate = 1

n = int(len(x))
iterations = 200

# Store all (m, c) after each iteration
history = []

# Step-2:
for i in range(iterations):
    y_prediction = m*x + c
    derivatives_m = (-2/n) * sum(x*(y-y_prediction))
    derivatives_c = (-2/n) * sum(y-y_prediction)

    # Step-3:
    m =  m - Learning_rate * derivatives_m
    c =  c - Learning_rate * derivatives_c
    cost = compute_cost(x, y, m, c)
    history.append((m, c, cost))  # Save the values for plotting

# Extract values for plotting
m_vals = [h[0] for h in history]
c_vals = [h[1] for h in history]
cost_vals = [h[2] for h in history]

# Plot
plt.figure(figsize=(10, 5))

# Plot 1: Gradient Descent Line Progression
plt.subplot(1, 3, 1)
plt.scatter(x, y, color='blue', alpha=0.5, label='Data Points')
plt.plot(x, y_pred, color='green', linewidth=2, label='Best Fit Line')
for i in range(iterations):
    y_pred = m_vals[i] * x + c_vals[i]
    if i == iterations - 1:
        plt.plot(x, y_pred, color='red', linewidth=2, label='Final Fit')
    else:
        plt.plot(x, y_pred, color='orange', alpha=0.3)
plt.title("Gradient Descent Line Progression")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Plot 2: Cost vs m
plt.subplot(1, 3, 2)
plt.plot(m_vals, cost_vals, marker='o', color='green')
plt.title("Cost vs m")
plt.xlabel("m")
plt.ylabel("Cost")
plt.grid(True)

# Plot 3: Cost vs b
plt.subplot(1, 3, 3)
plt.plot(c_vals, cost_vals, marker='o', color='purple')
plt.title("Cost vs b")
plt.xlabel("b")
plt.ylabel("Cost")
plt.grid(True)

plt.tight_layout()
plt.show()
```
