```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
actual    = np.array(['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog'])
predicted = np.array(['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
```

```python
print(accuracy_score(actual,predicted))
print(precision_score(actual,predicted,average=None))
print(recall_score(actual,predicted,average=None))
print(f1_score(actual,predicted,average=None))
```

```python
cm = confusion_matrix(actual,predicted)
cm
```

```python
sns.heatmap(cm, 
            annot=True,
            xticklabels=['Dog','Not Dog'],
            yticklabels=['Dog','Not Dog'])
plt.title('Confusion Matrix')
plt.xlabel('Prediction')
plt.ylabel('Actual')

plt.show()
```

```python
print(classification_report(actual, predicted))
```

#### ***Confusion Matrix For Multi-class Classification (Iris Datasets)***

```python
actual = ['Cat'] * 10 + ['Dog'] * 12 + ['Horse'] * 10
predicted = ['Cat'] * 8 + ['Dog'] + ['Horse'] + ['Cat'] * 2 + ['Dog'] * 10 + ['Horse'] * 8 + ['Dog'] * 2
classes = ['Cat', 'Dog', 'Horse']
```

```python
print(accuracy_score(actual,predicted))
print(precision_score(actual,predicted,average=None))
print(recall_score(actual,predicted,average=None))
print(f1_score(actual,predicted,average=None))
```

```python
print(classification_report(actual, predicted))
```

```python
cm = confusion_matrix(actual, predicted)
sns.heatmap(cm, 
            annot=True,
            xticklabels=classes,
            yticklabels=classes,
            cmap="Blues")
plt.title('Confusion Matrix')
plt.xlabel('Prediction')
plt.ylabel('Actual')

plt.show()
```
