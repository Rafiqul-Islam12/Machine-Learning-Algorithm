2. **sepal width in cm**
3. **petal length in cm**
4. **petal width in cm**

**class:**  
1. **Iris Setosa**

### **Visualisation - Scatterplot**

```python
colors = ['red', 'orange', 'blue']
species = iris["species"].unique()

plt.figure(figsize=(10, 6))

# 1. Sepal Length vs Sepal Width
plt.subplot(2, 2, 1)
plt.title('Sepal Dimensions')
sns.scatterplot(x=iris["sepal length (cm)"], 
                y=iris["sepal width (cm)"], 
                hue=iris["species"],palette = ["green","orange","blue"],
                s=100)

# 2. Petal Length vs Petal Width
plt.subplot(2, 2, 2)
plt.title('Petal Dimensions')
sns.scatterplot(x=iris["petal length (cm)"], 
                y=iris["petal width (cm)"], 
                hue=iris["species"],palette = ["green","orange","blue"],
                s=100)

# 3. Sepal Length vs Petal Length
plt.subplot(2, 2, 3)
plt.title('Sepal length vs Petal length Dimensions')
sns.scatterplot(x=iris["sepal length (cm)"], 
                y=iris["petal length (cm)"], 
                hue=iris["species"],palette = ["green","orange","blue"],
                s=100)

# 4. Sepal Width vs Petal Width
plt.subplot(2, 2, 4)
plt.title('Sepal Width vs Petal Width Dimensions')
sns.scatterplot(x=iris["sepal width (cm)"], 
                y=iris["petal width (cm)"], 
                hue=iris["species"],palette = ["green","orange","blue"],
                s=100)

plt.tight_layout()
plt.show()
```

### ***Model Training***

```python
from sklearn.model_selection import train_test_split
X = iris.drop(columns=['species'])
y = iris['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
tree.plot_tree(clf,filled=True,rounded=True,feature_names=X.columns)
```
